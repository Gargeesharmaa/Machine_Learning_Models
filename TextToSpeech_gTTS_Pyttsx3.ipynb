```python
from gtts import gTTS
import os

text = "hi iam a chatbot how can i help you"
text_to_speech = gTTS(text=text, lang='en')
text_to_speech.save('test.mp3')

#play the audio
os.system('test.mp3')
```

```python
TextToSpeech = pyttsx3.init()
voices = TextToSpeech.getProperty('voices')
TextToSpeech.setProperty('voice', voices[1].id)

rate = TextToSpeech.getProperty('rate')
TextToSpeech.setProperty('rate', 120)
TextToSpeech.say(text)
TextToSpeech.runAndWait()
TextToSpeech.save_to_file(text, 'test2.mp3')
```
